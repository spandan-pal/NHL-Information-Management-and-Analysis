```python
ratings=pd.read_csv('Ratings.csv')
ratings['HOME_TEAM']=ratings['HOME_TEAM'].str.replace('.', '',regex=False)
ratings.set_index('HOME_TEAM',inplace=True)
ratings.loc['FLA']
```

```python
df['TEAM_1']=df['TEAM_1'].str.replace('.', '',regex=False)
df['TEAM_2']=df['TEAM_2'].str.replace('.', '',regex=False)
```

```python
def rater(x):
    return ratings.loc[x]['Rating']
def win_predictor(x):
    c=x['H_Rating']-x['A_Rating']+ratings.loc['HT_ADV']['Rating']
    if c>0:
        return 'Yes'
    else:
        return 'No'
df['H_Rating']=df['TEAM_1'].apply(rater)
df['A_Rating']=df['TEAM_2'].apply(rater)
df['P_Win']=df.apply(win_predictor,axis=1)
```

```python
teams=df['TEAM_1'].append(df['TEAM_2'], ignore_index=True).drop_duplicates()
```

```python
def write_series_to_file():
    with open('series.txt', 'w') as file:
        file.write('match (n) detach delete n;'+'\n')
        file.write('CREATE'+'\n')
        for element in teams:
            file.write('('+str(element)+':TEAM{name:"'+str(element) +'"}),'+ '\n')
        for index, row in df.iterrows():
            file.write('('+str(row['TEAM_1'])+')-[:VS {Matches: '+str(row['NO_OF_MATCHES'])+',Wins: '+str(row['TEAM_1_WIN_COUNT'])+',Losses: '+str(row['TEAM_2_WIN_COUNT'])+',OT_Wins: '+str(row['TEAM_1_OVERTIME_WINS'])+',OT_Losses: '+str(row['TEAM_2_OVERTIME_WINS'])+',Playoff_Wins: '+str(row['TEAM_1_PLAYOFF_WINS'])+',Playoff_Losses: '+str(row['TEAM_2_PLAYOFF_WINS'])+',Avg_Goals_Scored: '+str(row['TEAM_1_AVG_GOALS_PER_MATCH'])+',Avg_Goals_Taken: '+str(row['TEAM_2_AVG_GOALS_PER_MATCH'])+',Our_Rating: '+str(row['H_Rating'])+',Opponent_Rating: '+str(row['A_Rating'])+',Predicted_to_Win: "'+str(row['P_Win'])+'"}]->('+str(row['TEAM_2'])+'),'+'\n')
        file.write(';')
```

```python
write_series_to_file()
with open('series.txt', 'r') as file:
    content = file.read()

# Truncate the file and write modified content (excluding the last character)
with open('series.txt', 'w') as file:
    file.write(content[:-3]+';')
```
